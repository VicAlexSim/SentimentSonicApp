```python
# dataset overview
import os
crema_files = os.listdir('../data/raw/crema-d')
iemocap_sessions = os.listdir('../data/raw/iemocap/session1')  # Adjust path if needed
print("CREMA-D files:", len(crema_files))
print("IEMOCAP sessions:", iemocap_sessions)
```

```python
# label distribution
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

labels = np.load('../data/processed/labels.npy', allow_pickle=True)
label_counts = pd.Series(labels).value_counts()
sns.barplot(x=label_counts.index, y=label_counts.values)
plt.title("Emotion Label Distribution")
plt.show()
```

```python
# audio duration distribution
import librosa

durations = [librosa.get_duration(filename='../data/raw/crema-d/' + f) for f in crema_files]
plt.hist(durations, bins=30)
plt.title("CREMA-D Audio Duration Distribution")
plt.show()
```

```python
# audio example waveform of a random crema-d file
import librosa.display

y, sr = librosa.load('../data/raw/crema-d/1001_DFA_ANG_XX.wav')
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform Example')
plt.show()
```
