```python
import numpy as np

# Load saved feature data
mfccs = np.load("../data/features/mfccs.npy")
spectrograms = np.load("../data/features/spectrograms.npy")
labels = np.load("../data/features/feature_labels.npy")

print("MFCCs shape:", mfccs.shape)
print("Spectrograms shape:", spectrograms.shape)
print("Labels shape:", labels.shape)
```

```python
import matplotlib.pyplot as plt
import librosa.display
import random

idx = random.randint(0, len(mfccs) - 1)

plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs[idx], x_axis='time')
plt.title(f"MFCC - Label: {labels[idx]}")
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(spectrograms[idx], x_axis='time', y_axis='mel', sr=16000)
plt.title(f"Mel Spectrogram - Label: {labels[idx]}")
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

label_counts = pd.Series(labels).value_counts()
print(label_counts)

label_counts.plot(kind='bar', title="Label Distribution")
plt.ylabel("Number of segments")
plt.xlabel("Emotion")
plt.show()
```
